```python

import pandas as pd
history = voo.history(period="max", auto_adjust=True)


year_end_prices = history['Close'].resample('YE').last()
yearly_returns = year_end_prices.pct_change() * 100
total_df = pd.DataFrame({
    'Year End Price': year_end_prices,
    'Yearly Returns': yearly_returns
})\
    
```
